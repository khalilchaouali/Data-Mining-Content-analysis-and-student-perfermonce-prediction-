```
import numpy as np
import pandas as pd
```

```
assessements=pd.read_csv('assessements_cleaned.csv')
cp_scores=pd.read_csv('checkpoint_scores_durations_cleaned.csv')
oto=pd.read_csv('one_to_one_cleaned.csv')
```

```
occ_cp=cp_scores['student id'].value_counts(sort=False).to_dict()
```

```
cp_scores['checkpoints mean']=cp_scores['student id']
cp_scores['checkpoints mean']=cp_scores.drop(columns='student id').replace(occ_cp)['checkpoints mean']
cp_scores_grouped=cp_scores.groupby(['student id']).mean()
```

```
occ=assessements['student id'].value_counts(sort=False).to_dict()
```

```
assessements['assessements mean']=assessements['student id']
assessements['assessements mean']=assessements.drop(columns='student id').replace(occ)['assessements mean']
```

```
assessements_grouped=assessements.groupby(['student id']).mean()
```

```
dataset = pd.merge(assessements_grouped, cp_scores_grouped,right_index=True, left_index=True)
```
